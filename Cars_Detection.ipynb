```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data="/content/car-detection-2/data.yaml",

    epochs=80,
    imgsz=512,
    batch=32,

    lr0=0.01,               # learning rate
    lrf=0.01,               # الموديل بيبدأ بـ lr0 ويقل تدريجيًا لحد lr0 * lrf
    momentum=0.937,         # يسرّع التعلم في الاتجاه الصحيح و بيمنع التذبذب
    weight_decay=0.0005,    # regularization بيحفظ الموديل يختار ويتس قليله
    warmup_epochs=3,        # يبدأ ببطء و يزود تدريجي
    patience=10,            # early stopping
)
```
